# Strum-Liouville theory

### ODE notation is

$
\large
\frac{d}{dx}\left[p(x) \frac{dy}{dx}
\right] + q(x)y = -\lambda w(x)y,
$

### example

$
\large
x^3 y^{\prime \prime} - x y^{\prime} + 2y = 0,
$

$
\large
y^{\prime \prime} - \frac{1}{x^2}y^\prime + \frac{2}{x^3}y = 0,
$

$
\large
\mu(x) = \exp \left( \int - \frac{dx}{x^2} \right) = e^{1/x},
$

$
\large
\frac{d}{dx}e^{1/x} = e^{1/x} \frac{d}{dx} \left( \frac{1}{x} \right) = - \frac{e^{1/x}}{x^2},
$

$
\large
\left( e^{1/x} y^\prime \right) = \frac{d}{dx} e^{1/x} y^\prime + e^{1/x} \frac{d}{dx} y^\prime
= - \frac{e^{1/x}}{x^2} y^\prime + e^{1/x}y^{\prime \prime},
$

$
\large
\therefore \left(e^{1/x} y^\prime \right)^\prime + \frac{2 e^{1/x}}{x^3} y = 0.
$

### We can make a generalization to that,

$
\large
P(x) y^{\prime \prime} + Q(x) y^\prime + R(x) y = 0,
$

$
\large
y^{\prime \prime} + \frac{Q(x)}{P(x)} y^\prime + \frac{R(x)}{P(x)} y = 0,
$

$
\large
\mu(x) = \frac{1}{P(x)} exp \left( \int \frac{Q(x)}{P(x)} dx \right),
$

$
\large
\therefore \left( \mu(x) P(x) y^\prime \right)^\prime + \mu (x) R(x) y = 0,
$

### explicity is of

$
\large
\frac{d}{dx} \left[ \exp \left( \int \frac{Q(x)}{P(x)} dx \right) y^\prime \right] + \frac{R(x)}{P(x)} \exp \left( \int \frac{Q(x)}{P(x)} dx \right) y = 0,
$

$
\large
\frac{d}{dx}\left[p(x) \frac{dy}{dx}
\right] + q(x)y = -\lambda w(x)y,
$

$
\large
\lambda y = -\frac{1}{w(x)} \left\{
\frac{d}{dx}\left[p(x) \frac{dy}{dx}
\right] + q(x)y
\right\}.
$

### We study the harmonic oscillator as,

$
\large
\lambda y = -\frac{d^2}{dx^2} y,
$

$
\large
y^{\prime \prime} + \lambda y = 0,
$

### for example of the fifth oder Dirichlet boundary condition, 

$
\large
y^{\prime \prime}(x) + \lambda y(x) = 0, \ \ y(0) = 0, y(\pi) = 0,
$

```
sol = DSolve[{y''[x] + \[Lambda] y[x] == 0, y[0] == 0, y[\[Pi]] == 0},
   y[x], x]
```

```
eigfuns = 
 Table[y[x] /. 
     sol[[1]] //. {\[FormalN] -> i, \[Lambda] -> \[FormalN]^2} /. {C[
      1] -> 1}, {i, 5}]
```

```
Plot[Evaluate[eigfuns], {x, 0, Pi}]
```

### Next, for example of the fifth oder Neumann boundary condition,

```
sol = DSolve[{y''[x] + \[Lambda] y[x] == 0, y'[0] == 0, 
   y'[\[Pi]] == 0}, y[x], x]
```

```
eigfuns = 
 Table[y[x] /. 
     sol[[1]] //. {\[FormalN] -> i, \[Lambda] -> \[FormalN]^2} /. {C[
      1] -> 1}, {i, 5}]
```

```
Plot[Evaluate[eigfuns], {x, 0, Pi}]
```
